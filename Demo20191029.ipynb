## 考選部

```python
from selenium import webdriver
driver = webdriver.Chrome('/opt/local/common/chromedriver')
driver.get('https://wwwq.moex.gov.tw/exam/wFrmExamQandASearch.aspx')
```

```python
driver.find_element_by_id('ctl00_holderContent_ibtnFull').click()
```

```python
ele = driver.find_element_by_id('ctl00_holderContent_txtCategoryName').send_keys('會計')
```

```python
driver.find_element_by_id('ctl00_holderContent_btnQuery').click()
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(driver.page_source, 'lxml')
```

```python
level1 = ''
level2 = ''
ary = []
for ele in soup.select('td'):
    if ele.get('class') and ('level1' in ele.get('class')):
        level1 = ele.text.strip()
    elif ele.get('class') and ('level2' in ele.get('class')):
        level2 = ele.text.strip()
    else:
        if ele.select_one('table'):
            exam_title = ele.select_one('table .exam-title').text.strip()
            exam_answers = ele.select('table .exam-question-ans')
            exam_answer1, exam_answer2,exam_answer3 = '','',''
            if exam_answers[0].select_one('a').get('href'):
                exam_answer1 = exam_answers[0].select_one('a').get('href')
            if exam_answers[1].select_one('a').get('href'):
                exam_answer2 = exam_answers[1].select_one('a').get('href')
            if exam_answers[2].select_one('a').get('href'):
                exam_answer3 = exam_answers[2].select_one('a').get('href')
            ary.append({'level1': level1, 
                        'level2':level2, 
                        'exam':exam_title,
                       'a1': exam_answer1,
                       'a2': exam_answer2,
                       'a3': exam_answer3})
        
```

```python
import pandas
df = pandas.DataFrame(ary)
```
